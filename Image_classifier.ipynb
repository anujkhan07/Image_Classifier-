```python
import tensorflow as tf
import numpy as np

# Load the MNIST dataset with corrected variable names
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()

# Reshape and normalize x_train, saving to x_train_final
x_train_final = x_train.reshape(x_train.shape[0], -1)
x_train_final = x_train_final / 255.0

# Define the model
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(10, activation='sigmoid', input_shape=(784,)))
model.compile(optimizer='adam' , loss='sparse_categorical_crossentropy' , metrics=['accuracy'])

# Train the model using x_train_final
model.fit(x_train_final, y_train, epochs = 20)
```

```python
model.save_weights('./my_model.weights.h5')
```
